```python
import selenium
import random
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from tkinter import *
import time
import pandas as pd
from bs4 import BeautifulSoup
import re as re
from openpyxl import Workbook, load_workbook
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import tkinter as tk


def SignIn():
    
    UserNameSearch = driver.find_element(By.ID,"username")
    PasswordSearch = driver.find_element(By.ID,"password")
    UserNameSearch.send_keys("Enter your linkedin Email here")
    PasswordSearch.send_keys("LinkedIn account Password here")
    PasswordSearch.send_keys(Keys.RETURN)
    
def extractInfo():
    src = driver.page_source
    soup = BeautifulSoup(src, 'lxml')
    print("------------------------ PERSON'S INFO -----------------------")
    info_div = soup.find('div', {'class': 'mt2 relative'})
    person_name=info_div.find('h1',{'class':'text-heading-xlarge inline t-24 v-align-middle break-words'})
    personName=person_name.text
    print("Name :",personName)
    person_bio=info_div.find('div',{'class':'text-body-medium break-words'})
    print("BIO :",person_bio.text.strip())
    person_location=info_div.find('span',{'class':'text-body-small inline t-black--light break-words'})

    print("location : ",person_location.text.strip())
     
    return personName



 
    
def calculateScores():
    print("--------------------------------- CALCULATING SCORES ----------------------------------")

    try:
        for k in skills:
            match = re.match(r"([a-z]+)([0-9]+)", k, re.I)
            if match:
                item = match.groups()
                skills[skills.index(k)] = item[0]
        definedLanguages = ["html css", "react", "flutter",
                            "javascript",  "java", "express", "node"]
        for b in skills:
            skills[skills.index(b)]=b.replace("."," ")
            skills[skills.index(b)]=b.replace("js","")
            for s in definedLanguages:
                if (b.lower() in s.lower()) or (s.lower() in b.lower()):
                    print(s,"------------------------",b)
                    if(s.lower()==b.lower()):
                        dictionery[s.replace(" ","")] = 1
                        print(s,"----------> SCORE ADDED")


                    if(len(b) < len(s)):
                         
                        word = s.split()
                         
                        for i in word:
                            if(i == b.lower()):

                                dictionery[s.replace(" ","")] = 1
                                print(s,"----------> SCORE ADDED")
                    if(len(b) > len(s)):
                         
                        word = b.split()
                         
                        for i in word:
                            if(i.lower() == s):

                                dictionery[s.replace(" ","")] = 1 
                                print(s,"----------> SCORE ADDED")
        #-----------------------------------------------------------------------------------------------------------------


        
        
    except ValueError as error:
        print()
    
    print("calculated scores: ", dictionery)

def makeExcelSheet():
    wb = Workbook()
    ws = wb.active

    ws.title = "Data"
    ws.append(['Names', 'HTML/CSS', 'React', 'Flutter','Javascript','Java','Express','Node','LinkedIn Profile'])
    wb.save('Data.xlsx')
    
    


    
def scroll():
    start = time.time()

    # will be used in the while loop
    initialScroll = 0
    finalScroll = 1000
    while True:
        driver.execute_script(
            f"window.scrollTo({initialScroll},{finalScroll})")
        initialScroll = finalScroll
        finalScroll += 1000
        time.sleep(3)
        end = time.time()
        if round(end-start) > 20:
            break

# objective function to evaluate the fitness of an individual
def evaluate_fitness(scores):
  return sum(scores)

# crossover operator to create two offspring from two parents
def crossover(p1, p2):
  # select a random crossover point
  crossover_point = random.randint(1, len(p1)-1)
  # create the offspring by combining the two parents at the crossover point
  c1 = p1[:crossover_point] + p2[crossover_point:]
  c2 = p2[:crossover_point] + p1[crossover_point:]
  return c1, c2

# mutation operator to randomly flip the scores of an individual
def mutation(scores, mutation_rate):
  for i in range(len(scores)):
    # flip the score with a probability of mutation_rate
    if random.random() < mutation_rate:
      scores[i] = 1 - scores[i]

# genetic algorithm function
def genetic_algorithm(scores_list, n_iter, mutation_rate):
  # initialize the population with the input scores
  population = scores_list
  # keep track of the best individual found
  best_individual = scores_list[0]
  best_fitness = evaluate_fitness(scores_list[0])
  # enumerate the generations
  for _ in range(n_iter):
    # create the offspring for the next generation
    children = []
    for i in range(0, len(population), 2):
      # select two parents
      p1, p2 = population[i], population[i+1]
      # crossover and mutation to create the offspring
      c1, c2 = crossover(p1, p2)
      mutation(c1, mutation_rate)
      mutation(c2, mutation_rate)
      # add the offspring to the population
      children.extend([c1, c2])
    # replace the population with the offspring
    population = children
    # check if any of the individuals in the population is better than the best individual found so far
    for individual in population:
      fitness = evaluate_fitness(individual)
      if fitness > best_fitness:
        best_individual = individual
        best_fitness = fitness
  return best_individual            
            

        

profile_url=""
profile_url_1=""
candidates=0
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
skills=[]
dictionery = {"htmlcss": 0, "react": 0, "flutter": 0, "javascript": 0, "java": 0, "express": 0, "node": 0}   
profiles=[]  
result=[]
def candidateFunction():
    global name_profile
    global HTML_CSS_Skill
    global React_Skill
    global flutter_Skill
    global javascript_Skill
    
    global java_Skill
    global express_Skill
    global node_Skill
    global profile_linkedin
    l3=Label(text="Name= "+ name_profile)
    l3.pack()
    l4=Label(text="HTML/CSS Score = " + str(HTML_CSS_Skill))
    l4.pack()
    l5=Label(text="React Skill Score = " + str(React_Skill))
    l5.pack()
    l6=Label(text="Flutter Score = " + str(flutter_Skill))
    l6.pack()
    l7=Label(text="Javascript Skill = " + str(javascript_Skill))
    l7.pack()
    
    
    l9=Label(text="Java Score = " + str(java_Skill))
    l9.pack()
    l10=Label(text="Express Score = " + str(express_Skill))
    l10.pack()
    l11=Label(text="Node Skill = " + str(node_Skill))
    l11.pack()
    l12=Label(text="LinkedIn Profile = " + profile_linkedin)
    l12.pack()
def aiFunction():
    driver.get("https://www.linkedin.com/login")
    SignIn()  
    makeExcelSheet()
    for i in profiles:

        linkedin=i
        profile_url=i
        driver.get(profile_url)
        scroll()

        name = extractInfo()




        profile_url_1 = profile_url+"details/skills/"
        driver.get(profile_url_1)
        scroll()

        src1 = driver.page_source

        soup1 = BeautifulSoup(src1, 'lxml')
        try:

            skill_div = soup1.find('main', {'class': 'scaffold-layout__main'})
            s1 = skill_div.find('section', {'class': 'artdeco-card ember-view pb3'})
            s2 = s1.find('div', {'class': 'artdeco-tabs artdeco-tabs--size-t-48 ember-view'})
            s3 = s2.find('div', {'class': 'artdeco-tabpanel active ember-view'})
            s4 = s3.find('div', {'class': 'pvs-list__container'})
            skill_div_1 = s4.find('div', {'class': 'scaffold-finite-scroll__content'})

            ul_tag=skill_div_1.find('ul',{'class':'pvs-list'})
            li_tag = ul_tag.find_all('li')
            for i in li_tag:
                try:
                    firstdiv1 = i.find('div',{'class':"pvs-entity pvs-entity--padded pvs-list__item--no-padding-when-nested"})
                    secdiv1 = firstdiv1.find('div',{'class':"display-flex flex-column full-width align-self-center"})
                    thirdiv1 = secdiv1.find('div',{'class':"display-flex flex-row justify-space-between"})
                    atag1=thirdiv1.find('a',{'class':"optional-action-target-wrapper display-flex flex-column full-width"})
                    fourdiv1 = thirdiv1.find('div',{'class':"display-flex align-items-center"})
                    firstspan1 = fourdiv1.find('span',{'class':"mr1 hoverable-link-text t-bold"})
                    skill = firstspan1.find('span', attrs={'aria-hidden': "true"}).text.strip()
                    skills.append(skill)
                except AttributeError as error:
                        print()
        except AttributeError as error:
                        print()



        calculateScores()

        wb =load_workbook(r'C:\Users\Muhammad Uzair\Data.xlsx')
        ws = wb.active   
        newList=list(dictionery.values())
        print(newList)
        result.append(list(dictionery.values()))
        newList.insert(0,name)
        newList.append(linkedin)
        ws.append(newList)
        wb.save(r'C:\Users\Muhammad Uzair\Data.xlsx')  





    print("-------------------Array of Scores----------------------------")
    print(result)
    # test the genetic algorithm
    scores_list = result
    fittest_individual = genetic_algorithm(scores_list, 1000, 0.1)
    print(fittest_individual)
    input_array = fittest_individual
    df = pd.read_excel('Data.xlsx')
    
    data_list = df.values.tolist()
    global recommended_candidate
    global name_profile
    global HTML_CSS_Skill
    global React_Skill
    global flutter_Skill
    global javascript_Skill
    
    global java_Skill
    global express_Skill
    global node_Skill
    global profile_linkedin
    flag=False
    for i in data_list:

        if ((i[1]==input_array[0]) and (i[2]==input_array[1]) and (i[3]==input_array[2]) and (i[4]==input_array[3]) and (i[5]==input_array[4]) and (i[6]==input_array[5]) and (i[7]==input_array[6]) ):
            flag=True
            recommended_candidate=i
            break 

    if (flag==False):
        print("Your level of judgement is high :(  ")


    print(recommended_candidate)
    name_profile=recommended_candidate[0]
    HTML_CSS_Skill=recommended_candidate[1]
    React_Skill=recommended_candidate[2]
    flutter_Skill=recommended_candidate[3]
    javascript_Skill=recommended_candidate[4]
    
    java_Skill=recommended_candidate[5]
    express_Skill=recommended_candidate[6]
    node_Skill=recommended_candidate[7]
    profile_linkedin=recommended_candidate[8]
    print("Recommended Candidate: ")
    print("Name: ",name_profile)
    print("HTML/CSS: ",HTML_CSS_Skill)
    print("React: ",React_Skill)
    print("Flutter: ",flutter_Skill)
    print("Javascript: ",javascript_Skill)
    
    print("Java: ",java_Skill)
    print("Express: ",express_Skill)
    print("Node: ",node_Skill)
    print("LinkedIn Profile: ",profile_linkedin)
    candidateFunction()
def main():
    print("-------------------Smart recruitment System-----------------------")
    
    def counterFunction():
        global candidates

        global profile_url
        global profile_url_1

        global skills
        global result
        
        candidates=candidates+1
        input_text=textField.get()
        profiles.append(input_text)
        scValue.set("")
        print(profiles)
        l2.config(text="Selected Candidates= "+str(candidates))

        print(candidates)

    

    App_root = Tk()
    # gui logic here
    # Width X Height
    App_root.geometry("700x400")
    App_root.minsize(700, 400)
    App_root.maxsize(700, 400)
    App_root.title("Smart Recuritment System")
    candidates=0
    candidatesStr=str(candidates)


    l1 = Label(text="Smart Recuritment System",font=150,pady=10)
    l1.pack()
    f = Frame(App_root,width=100)

    scValue = StringVar()
    scValue.set("")
    textField = Entry( f,textvariable=scValue,font="lucida 15")
    textField.pack(side="left", fill=X,pady=10)
    b1 = Button(f,  fg="white", text="Add Candidate", bg="#008080", command=counterFunction, width=30)
    b1.pack(side="left")
    f.pack()

    l2=Label(text="Selected Cendidiates= "+str(candidates))
    l2.pack()
    
    b1 = Button(App_root,  fg="white", text="Show the Best Candidate", bg="#008080", command=aiFunction, width=50)
    b1.pack()
    
    App_root.mainloop()

if __name__=='__main__':
    main()













```
